```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
iris_dataset = load_iris()
print("Iris Features/Target Names:",iris_dataset.target_names)
```

```python
for i in range(len(iris_dataset.target_names)):
  print("\n[{0}] [{1}]:".format(i,iris_dataset.target_names[i]))
print("Iris data",iris_dataset["data"])
x_train,x_test,y_train,y_test = train_test_split(iris_dataset["data"],iris_dataset['target'],random_state=0)

print('Targets : {}'.format(iris_dataset['target']))
print('X Train:',x_train.shape)
print('X Test:',x_test.shape)
print(' Y Train:',y_train.shape)
print(' Y Test :',y_test.shape)
kn = KNeighborsClassifier(n_neighbors=1)
kn.fit(x_train,y_train)

x_new = np.array([[5,2.9,1,0.2]])
print("X_new is :",x_new)

prediction = kn.predict(x_new)

print('Predicted target names {}'.format(prediction))
print("Predicted feature name {}".format(iris_dataset['target_names'][prediction]))

i=1
x = x_test[i]
x_new = np.array([x])

for i in range(len(x_test)):
  x = x_test[i]
  x_new = np.array([x])
  prediction = kn.predict(x_new)
  print("Actual : {0} {1}, Predicted : {2} {3}".format(y_test[i],iris_dataset['target_names'][y_test[i]],prediction,iris_dataset['target_names'][prediction]))
print(" TEST SCORE[Acuracy] : {:.2f}".format(kn.score(x_test,y_test)))
```
